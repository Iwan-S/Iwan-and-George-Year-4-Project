```python
Titles = ['ID', r'T$_{eff}$ ($K$)', r'FeH', r'Δν ($μHz$)', r'ν$_{max}$ ($μHz$)',
          r'Radius ($R_{\odot}$)', r'Mass ($M_{\odot}$)', r'Age ($Myr$)']
```

```python
def ScreamExtractor(Folder, Head, Foot):
    return np.genfromtxt(f'{Directory}/{Folder}/results.ascii', dtype='str',
                         comments='COMMENT', skip_header=Head, skip_footer=Foot)
```

```python
def Errors(Folder, Error):
    return [[float(Folder[Error])], [float(Folder[Error+1])]]
```

```python
Global = np.array([ScreamExtractor('Global', 0, 1)[1:], ScreamExtractor('Global', 1, 0)])
Frequency = np.array([ScreamExtractor('Freqs', 0, 1)[1:], ScreamExtractor('Freqs', 1, 0)])
Epsilon = np.array([ScreamExtractor('Epsilon', 0, 1)[1:], ScreamExtractor('Epsilon', 1, 0)])
json = np.array([ScreamExtractor('json', 0, 1)[1:], ScreamExtractor('json', 1, 0)])
Ratio = np.array([ScreamExtractor('Ratio', 0, 1)[1:], ScreamExtractor('Ratio', 1, 0)])
```

```python
Cases = [Global, Frequency, Epsilon, json, Ratio]
CasesNames = ['Global', 'Frequency', 'Epsilon', 'json', 'Ratio']
```

```python
fig, [ax1, ax2, ax3, ax4, ax5, ax6, ax7] = plt.subplots(1, 7, sharey=True)
Plots = [ax1, ax2, ax3, ax4, ax5, ax6, ax7]
for m in range(5):
    for n in range(7):
        Plots[n].errorbar(float(Cases[m][1][3*n+1]), m+1, xerr=Errors(Cases[m][1], 3*n+2), fmt='*')
        Plots[n].set_xlabel(Titles[n+1])

plt.yticks(ticks=np.linspace(1, 5, 5), labels=CasesNames)

plt.show()
```

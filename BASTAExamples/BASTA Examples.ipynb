```python
Directory = 'C:/Users/georg/Documents/GitHub/Iwan-and-George-Year-4-Project/BASTAExamples'
```

```python
def ScreamExtractor(Folder):
    return np.genfromtxt(f'{Directory}/{Folder}/results.ascii', dtype='str')
```

```python
def Errors(Folder):
    return [[float(Folder[1][2])], [float(Folder[1][3])]]
```

```python
Global = ScreamExtractor('Global')
Freqs = ScreamExtractor('Freqs')
```

```python
plt.figure()
plt.bar(0, float(Global[1][1]), yerr=Errors(Global), color='blue')
plt.bar(0.8, float(Freqs[1][1]), yerr=Errors(Freqs), color='green')
plt.show()
```

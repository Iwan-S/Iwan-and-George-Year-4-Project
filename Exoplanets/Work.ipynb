```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
def ScreamExtractor(File):
    FILENAME = pd.read_csv(f'{File}.csv', sep=',', comment='#')
    return FILENAME
```

```python
Data = ScreamExtractor(f'All Stars Data')
```

```python
TempGaia = Data['Temperature (Gaia)']
TempExo = Data['Temperature (Exo)']
MetalGaia = Data['Metalicity (Gaia)']
MetalExo = Data['Metalicity (Exo)']
```
